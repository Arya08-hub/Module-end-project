```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
file_path ="C:\\Users\\aryas\\Downloads\\myexcel.xlsx"
xlsx = pd.ExcelFile(file_path)

# Display sheet names
xlsx.sheet_names
# Load the data from the sheet
df = pd.read_excel(xlsx, sheet_name="myexcel.csv")

# Display the first few rows to understand its structure
df.head()
import numpy as np

# Replace the "Height" column with random values between 150 and 180
df["Height"] = np.random.randint(150, 181, size=len(df))

# Verify the changes
df.head()

# Calculate employee distribution across teams
team_distribution = df["Team"].value_counts()

# Calculate percentage split
team_percentage = (team_distribution / len(df)) * 100

# Combine both into a DataFrame
team_stats = pd.DataFrame({"Employee Count": team_distribution, "Percentage": team_percentage})

# Display the results
team_stats
import matplotlib.pyplot as plt

# Plot the employee distribution across teams
plt.figure(figsize=(12, 6))
team_stats["Employee Count"].plot(kind="bar", color="skyblue", edgecolor="black")

# Labels and title
plt.xlabel("Teams")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
# Count employees by position
position_distribution = df["Position"].value_counts()

# Convert to DataFrame
position_stats = pd.DataFrame({"Employee Count": position_distribution})

# Display results
position_stats
# Plot position distribution as a pie chart
plt.figure(figsize=(8, 8))
plt.pie(position_stats["Employee Count"], labels=position_stats.index, autopct="%1.1f%%", colors=["lightblue", "lightgreen", "lightcoral", "lightskyblue", "gold"], startangle=140)

# Title
plt.title("Employee Distribution by Position")

# Show the plot
plt.show()
# Define age bins and labels
age_bins = [20, 25, 30, 35, 40, 45]
age_labels = ["20-24", "25-29", "30-34", "35-39", "40-44"]

# Categorize employees into age groups
df["Age Group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=False)

# Count employees in each age group
age_distribution = df["Age Group"].value_counts().sort_index()

# Convert to DataFrame
age_stats = pd.DataFrame({"Employee Count": age_distribution})

# Display results
age_stats
# Plot age group distribution as a bar chart
plt.figure(figsize=(8, 5))
age_stats["Employee Count"].plot(kind="bar", color="lightcoral", edgecolor="black")

# Labels and title
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution by Age Group")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
# Group by Team and sum salaries
team_salary_expense = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)

# Group by Position and sum salaries
position_salary_expense = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)

# Get the highest salary expenditure team and position
highest_salary_team = team_salary_expense.idxmax()
highest_salary_position = position_salary_expense.idxmax()
highest_salary_team, highest_salary_position
```
